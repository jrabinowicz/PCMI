# Experimento 1: Heurísticas

## Heuristica 1



```python
# Graficamos los tiempos de ejecución de cada dataset en función de n.
df_h1 = df_resultadosHeuristicas[df_resultadosHeuristicas["dataset"]=="H1"].copy();
df_h1["tiempo_esperado"] = 0.00006 * df_h1["n"] ** 3; # Exponencial
sns.scatterplot(data=df_h1, x="n", y="tiempo", s=80)
fig = sns.lineplot(data=df_h1, x='n', y='tiempo_esperado', color="orange");
fig.legend(labels=['O(n^3)','H1'])
fig.set(xlabel='Tamaño de la entrada', ylabel='Tiempo (ms)');
plt.savefig("graficos/h1.svg");
```

## Heurística 2

```python
# Graficamos los tiempos de ejecución de cada dataset en función de n.
df_h2= df_resultadosHeuristicas[df_resultadosHeuristicas["dataset"]=="H2"].copy();
df_h2["tiempo_esperado"] = 0.00059 * df_h2["n"] ** 3; # Exponencial
sns.scatterplot(data=df_h2, x="n", y="tiempo", s=80)
fig = sns.lineplot(data=df_h2, x='n', y='tiempo_esperado', color="orange");
fig.legend(labels=['O(n^3)','H2'])
fig.set(xlabel='Tamaño de la entrada', ylabel='Tiempo (ms)');
plt.savefig("graficos/h2.svg");
```

## Comparación del tiempo de ejecución

```python
# Graficamos los tiempos de ejecución de cada dataset en función de n.
df_tiempos = df_resultadosHeuristicas.copy();
sns.scatterplot(data=df_tiempos, x="n", y="tiempo", s=80, hue="dataset")
fig.set(xlabel='Tamaño de la entrada', ylabel='Tiempo (ms)');
plt.savefig("graficos/heuristicas-tiempo.svg");
```

## Comparación de GAP

```python
df_resultadosHeuristicas = df_resultadosHeuristicas.copy()
df_resultadosHeuristicas["gap"] = (df_resultadosHeuristicas["Impacto Optimo"] - df_resultadosHeuristicas["Impacto Conseguido"])/df_resultados["Impacto Optimo"];
df_heatmap = df_resultadosHeuristicas.pivot_table(index='n', columns='dataset', values='gap', aggfunc=np.mean);
fig = sns.heatmap(df_heatmap, cmap="Blues");
plt.savefig("graficos/heuristicas-heatmap.svg");
```

## Calidad de las soluciones generadas

```python
df_calidad = df_resultados.copy()
df_calidad = df_calidad[~((df_calidad["dataset"]=="H1")|(df_calidad["dataset"]=="H2"))]
df_calidad["gap"] = (df_calidad["Impacto Optimo"] - df_calidad["Impacto Conseguido"])/df_calidad["Impacto Optimo"];

############################################################################

df_calidad_h1 = df_calidad[     (df_calidad["dataset"]=="TB-H1-ME-MBI")|
                                (df_calidad["dataset"]=="TB-H1-ME-BI")|
                                (df_calidad["dataset"]=="TB-H1-ME-MI")|
                                (df_calidad["dataset"]=="TB-H1-ME-AI")|
                                (df_calidad["dataset"]=="TB-H1-MSE-MBI")|
                                (df_calidad["dataset"]=="TB-H1-MSE-BI")|
                                (df_calidad["dataset"]=="TB-H1-MSE-MI")|
                                (df_calidad["dataset"]=="TB-H1-MSE-AI")]

df_calidad_h2 = df_calidad[     (df_calidad["dataset"]=="TB-H2-ME-MBI")|
                                (df_calidad["dataset"]=="TB-H2-ME-BI")|
                                (df_calidad["dataset"]=="TB-H2-ME-MI")|
                                (df_calidad["dataset"]=="TB-H2-ME-AI")|
                                (df_calidad["dataset"]=="TB-H2-MSE-MBI")|
                                (df_calidad["dataset"]=="TB-H2-MSE-BI")|
                                (df_calidad["dataset"]=="TB-H2-MSE-MI")|
                                (df_calidad["dataset"]=="TB-H2-MSE-AI")]
```

```python
df_heatmap = df_calidad_h1.pivot_table(index='n', columns='dataset', values='gap')
df_heatmap = df_heatmap[[   "TB-H1-ME-MBI",
                            "TB-H1-MSE-MBI",
                            "TB-H1-ME-BI",
                            "TB-H1-MSE-BI",
                            "TB-H1-ME-MI",
                            "TB-H1-MSE-MI",
                            "TB-H1-ME-AI",
                            "TB-H1-MSE-AI"]].copy()
fig = sns.heatmap(df_heatmap, cmap="Blues");
plt.savefig("graficos/tb-h1-heatmap.svg");
```
